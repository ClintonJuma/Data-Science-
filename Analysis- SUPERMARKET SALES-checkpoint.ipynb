```python
#Grouping
df
```

```python
#grouping by customer type
by_customer_type = df.groupby(by= [ "Customer type", "Branch", "City", "Product line"])["Branch"].count().reset_index(name = "count")
by_customer_type = by_customer_type.sort_values("count", ascending = False)
by_customer_type = by_customer_type.set_index("Customer type")
by_customer_type
```

```python
#Customer type with more then 30 customers
Customer_type_query = by_customer_type.query("`count`> 30")
Customer_type_query
```

```python
#group by Branch
by_Branch = df.groupby(by = ["Branch", "City", "Product line"])["Branch"].count().reset_index(name = "count")
by_Branch = by_Branch.set_index("Branch")
by_Branch = by_Branch.sort_values("count",ascending = False)
by_Branch
```

```python
#query of branch with more thn fifty customers
Branch_query = by_Branch.query("`count`>50")
Branch_query
```

```python
#groupby product line
by_Product_line = df.groupby(by = ["Product line","Branch", "Gender"])["Branch"].count().reset_index(name = "count")
by_Product_line = by_Product_line.set_index("Product line")
by_Product_line = by_Product_line.sort_values("count", ascending = False)
by_Product_line
```

```python
#product line query on product line with more than 30 clients
Product_line_query = by_Product_line.query("`count`>30")
Product_line_query
```

```python
#grouping  y Payment method
by_Payment = df.groupby(by = ["Payment", "Branch", "Product line"])["Payment"].count().reset_index(name = "count")
by_Payment = by_Payment.set_index("Payment")
by_Payment = by_Payment.sort_values("count", ascending = False)
by_Payment                                 
                
```

```python
#Paymet query for payment method used by more than 20 clients
Payment_query = by_Payment.query("`count`>20")
Payment_query
```

```python
# grouping by dates sold
dates_sold = df.groupby(by = ["Day_sold", "Month_sold", "City", "Product line", "Branch"])["City"].count().reset_index(name = "count")
dates_sold = dates_sold.sort_values("count", ascending =False)
dates_sold = dates_sold.set_index("count")
dates_sold
```

```python
#grouping by Gender
by_Gender = df.groupby(by = ["Gender", "Customer type", "Branch", "Product line"])["Gender"].count().reset_index(name = "count")
by_Gender = by_Gender.set_index("Gender")
by_Gender = by_Gender.sort_values("count", ascending = False)
by_Gender
```

```python
#Gender query with mre than 10 customers
Gender_query = by_Gender.query("`count`>10")
Gender_query
```

```python

#grouping by Rating
by_Rating  =df.groupby(by = ["Rating", "Product line"])["Rating"].count().reset_index(name="count")
by_Rating = by_Rating.sort_values("Rating", ascending =False)
by_Rating = by_Rating.set_index("Rating")
by_Rating
```

```python
#Rating query for Ratings above 9.0
Rating_query = by_Rating.query("`Rating`>5.0")
Rating_query
```
