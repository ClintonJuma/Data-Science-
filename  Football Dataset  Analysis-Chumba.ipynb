```python
#grouping by goals, name , Club and position
player_goals=df.groupby(["Name", "Goals", "Club", "Position"])["Name"].count().reset_index(name="count")
player_goals=player_goals.set_index("Goals")
#sort
player_goals=player_goals.sort_values(by=["Goals"], ascending=False)
player_goals
```

```python
#grouping by Assists, Name, Club, Passes Attempted and  assists
player_assists=df.groupby(["Name", "Assists", "Club", "Passes_Attempted"])["Name"].count().reset_index(name="count")
player_assists=player_assists.set_index("Assists")
#sort
player_assists=player_assists.sort_values(by=["Assists"], ascending=False)
player_assists
```

```python
#grouping by Age, Name and Nationality
player_age=df.groupby(["Name", "Age", 'Club', "Nationality"])["Age"].count().reset_index(name="count")
player_age=player_age.sort_values(by=["Age"], ascending=False)
player_age=player_age.set_index("Age")
player_age
```

```python
#grouping by Yellow-Cards, Name and Club
yellowcard_players = df.groupby(["Name", "Club", "Yellow_Cards"])["Name"].count().reset_index(name="count")
yellowcard_players = yellowcard_players.sort_values(by=["Yellow_Cards"], ascending=False)
yellowcard_players = yellowcard_players.set_index("Yellow_Cards")
yellowcard_players
```

```python
#grouping by Matches, Name and Club
played_matches=df.groupby(["Name", "Matches", "Club"])["Matches"].count().reset_index(name="count")
played_matches=played_matches.sort_values(by=["Matches"], ascending=False)
played_matches=played_matches.set_index("Matches")
played_matches
```

```python
#query to find players with more than 17 goals
top_scorers=player_goals.query("`Goals` >=17")
top_scorers
```

```python
#query to find players with more then 10 assists
top_assists = player_assists.query("`Assists` >=10")
top_assists
```

```python
#query to find players above 30 years 
top30_yrs = player_age.query("`Age` >30")
top30_yrs
```

```python
#query to find players who played all 38 matches matches
total_matches = played_matches.query("`Matches` >=38")
total_matches
```
