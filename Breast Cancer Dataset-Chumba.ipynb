```python
#grouping by  Patient ID, Age ,Surgery type and date of surgery
PatientID_Age = df.groupby(by=["Patient_ID", "Age", "Surgery_type", "Date_of_Surgery"])["Age"].count().reset_index(name="count")
PatientID_Age = PatientID_Age.set_index("Patient_ID")
PatientID_Age = PatientID_Age.sort_values("Age", ascending=False)
PatientID_Age
```

```python
#grouping by Age, Surgery type and Histology
Age_Surgery_Histology = df.groupby(by=["Age", "Surgery_type", "Histology", "Tumour_Stage"])["Age"].count().reset_index(name="count")
Age_Surgery_Histology = Age_Surgery_Histology.set_index("Age")
Age_Surgery_Histology = Age_Surgery_Histology.sort_values("Age", ascending=False)
Age_Surgery_Histology
```

```python
#group by Patient_ID, Age and Tumour_Stage
Age_Tumour = df.groupby(by=["Patient_ID", "Age", "Tumour_Stage"])["Age"].count().reset_index(name="count")
Age_Tumour = Age_Tumour.set_index("Patient_ID")
Age_Tumour = Age_Tumour.sort_values("Age", ascending = False)
Age_Tumour
```

```python
#grouping by Tumour stage and Surgery type
Tumour_Surgery = df.groupby(by=["Surgery_type", "Tumour_Stage"])["Tumour_Stage"].count().reset_index(name="count")
Tumour_Surgery = Tumour_Surgery.set_index("count")
Tumour_Surgery = Tumour_Surgery.sort_values("count", ascending=False)
Tumour_Surgery
```

```python
#grouping Tumour stage and patient_status
Tumour_PatientStatus = df.groupby(by = ["Tumour_Stage", "Patient_Status"])["Tumour_Stage"].count().reset_index(name = "count")
Tumour_PatientStatus = Tumour_PatientStatus.set_index("count")
Tumour_PatientStatus
```

```python
#query to find Histology of patients above 80 years
Top80yrs_Histology = Age_Surgery_Histology.query("`Age` >= 80")
Top80yrs_Histology

```

Most patients above 80 years old haveInfiltrating Ductal Carcinoma

```python
#Query for Surgery_type and Tumour_Stage with more than 30 patients
Most_Patients = Tumour_Surgery.query("`count` >= 0")
Most_Patients
```

```python
#query  Tumour_Stage  and Patient status 
Status_TumourStage = Tumour_PatientStatus.query("`count` >= 0")
Status_TumourStage
```
